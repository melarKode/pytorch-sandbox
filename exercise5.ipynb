```python
import torch
import numpy as np
```

```python
t_c = torch.tensor([0.5, 14.0, 15.0, 28.0, 11.0, 8.0, 3.0, -4.0, 6.0, 13.0, 21.0])

t_u = torch.tensor([35.7, 55.9, 58.2, 81.9, 56.3, 48.9, 33.9, 21.8, 48.4, 60.4, 68.4])
```

```python
def model(t_u,w1, w2,b):
    return w2*t_u**2 + w1*t_u + b
```

```python
def loss_fn(t_p, t_c):
    return torch.mean((t_p - t_c)**2)
```

```python
import torch.optim as optim

params = torch.tensor([1.0,1.0,0.0],requires_grad=True)
learning_rate = 1e-2
optimizer = optim.Adam([params], lr=learning_rate)
```

```python
def training_loop(n_epochs, optimizer, t_u, t_c, params):
    for epoch in range(n_epochs):
        t_p = model(t_u, *params)
        loss = loss_fn(t_p, t_c)
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        print('Epoch {:3d}/{} Loss: {:.6f} Params: {}'.format(epoch, n_epochs, loss.item(), params.detach().numpy()))
```

```python
training_loop(3000, optimizer, t_u, t_c, params)
```
